```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
filename = 'D:/研究所/1132/智慧行銷管理/期末報告/Preprocessed_data.csv'
data = pd.read_csv(filename)
data = data.astype(int)
```

```python
buy = data[data['overall_purchase_intention_anessa'] == 1]
not_buy = data[data['overall_purchase_intention_anessa'] != 1]
```

```python
data_cols = data.columns
for col in data_cols:
    print(data[col].value_counts(sort = False))
    print(data[col].value_counts(sort = False, normalize = True).round(4))
    print('------------------------------------')
```

```python
for col in data_cols:
    print(buy[col].value_counts(sort = False))
    print(buy[col].value_counts(sort = False, normalize = True).round(4))
    print('------------------------------------')
```

```python
for col in data_cols:
    print(not_buy[col].value_counts(sort = False))
    print(not_buy[col].value_counts(sort = False, normalize = True).round(4))
    print('------------------------------------')
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 3, figsize = (18, 12))
axes = axes.flatten() 

columns_to_plot = ['monthly_spending', 'used_product', 'time_used',
                   'usage_frequency', 'most_frequently_used_product', 'anessa_products_used']

titles = ['Monthly Spending Distribution', 'Used Product Distribution', 'Time Used Distribution',
          'Usage Frequency Distribution', 'Most Frequently Used Product Distribution',
          'Anessa Products Used Distribution']

palettes = ['Blues', 'Reds', 'Purples', 'Greys', 'Oranges', 'Greens']

for i, col in enumerate(columns_to_plot):
    sns.countplot(data = data, x = col, palette = palettes[i], ax = axes[i])

    axes[i].set_title(titles[i])
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation = 45, ha = 'right')

plt.tight_layout()

plt.show()
```
