如果「請問您曾經使用過防曬商品嗎?」這一題不是「沒使用過」，在使用多久時間及頻率兩題不能為空值，如果是空值，則移除此筆資料

```python
condition = data['used_product'].str.contains('沒有使用過', na = False)
condition_B = data['time_used'].isna() | data['usage_frequency'].isna()
data = data[~(~condition & condition_B)]
```

```python
data.loc[condition, ['time_used', 'usage_frequency']] = np.nan
```
